# Q 06

Q. 다음의 그래프에 대해 Prim의 알고리즘을 이용해 MST를 구하라.

```python
INF = 9999

g = [
    [INF, 7, 4, INF, 2],
    [7, INF, INF, 3, 4],
    [4, INF, INF, 6, 3],
    [INF, 3, 6, INF, 5],
    [2, 4, 3, 5, INF]
]

vertex = ['A', 'B', 'C', 'D', 'E']

def MSTPrim(g, vertex):
    n = len(vertex)
    
    dist = [INF] * n
    dist[0] = 0 
    
    selected = [False] * n
    
    for i in range(n):
        u = getMinVertex(dist, selected)
        selected[u] = True
        
        print('선택된 정점, 간선의 가중치 >>', vertex[u], ',', dist[u], end = '   : ')
        
        for v in range(n):
            if not selected[v] and g[u][v] != None:
                dist[v] = min(g[u][v], dist[v])
                
        print(dist)
        
        
def getMinVertex(dist, selected):
    minv = -1
    mindist = INF
    
    for i in range(len(dist)):
        if not selected[i] and mindist > dist[i]:
            mindist = dist[i]
            minv = i
    
    return minv


MSTPrim(g, vertex)
```

# Q 12

Q. Q06의 그래프에 대해 Kruskal의 알고리즘을 이용해 MST를 구하라.

```python
INF = 9999

g = [
    [INF, 7, 4, INF, 2],
    [7, INF, INF, 3, 4],
    [4, INF, INF, 6, 3],
    [INF, 3, 6, INF, 5],
    [2, 4, 3, 5, INF]
]

vertex = ['A', 'B', 'C', 'D', 'E']

# 간선리스트. 가중치가 작은 간선부터 차례로 선택한다. 사이클이 생기지 않는 한에서

class disjointSets:
    def __init__(self, n):
        self.parent = [-1] * n
        
    def find(self, id):
        while self.parent[id] >= 0:
            id = self.parent[id]
            
        return id
    
    def union(self, s1, s2):
        self.parent[s1] = s2
        

def MSTKruskal(graph, vertex):
    n = len(vertex)
    
    dsets = disjointSets(n)
    
    E = []
    
    for i in range(n-1):
        for j in range(i+1, n):
            if graph[i][j] != INF:
                E.append((i, j, graph[i][j]))
                
    E.sort(key = lambda e : e[2])
    
    ecount = 0
    
    for e in E:
        uset = dsets.find(e[0])
        vset = dsets.find(e[1])
        
        if uset != vset:
            dsets.union(uset, vset)
            print("간선 추가 : (%s, %s, %d)" % (vertex[e[0]], vertex[e[1]], e[2]))
            ecount += 1
            if ecount == n-1:
                break
                
                
MSTKruskal(g, vertex)  
```

# Q 22

Q. 다음 그래프에 대해 Dijkstra 알고리즘을 이용해 정점 A에서 출발해 모든 정점으로 가는 최단경로와 그 길이를 구하는 과정을 보여라.

```python
INF = 9999

g = [
    [INF, 7, 9, INF, INF, 14],
    [7, INF, 10, 15, INF, INF],
    [9, 10, INF, 11, INF, INF],
    [INF, 15, 11, INF, 6, INF],
    [INF, INF, INF, 6, INF, 9],
    [14, INF, INF, INF, 9, INF]
]

vertex = ['A', 'B', 'C', 'D', 'E', 'F']

def dijkstra(graph, vertex, start):
    n = len(vertex)
    
    dist = list(graph[start])
    dist[start] = 0
    path = [start] * n
        
    selected = [False] * n
    selected[start] = True
    
    for i in range(n):
        u = getMinVertex(dist, selected)
        selected[u] = True
        
        for w in range(n):
            if not selected[w]:
                if dist[w] > dist[u] + graph[u][w]:
                    dist[w] = dist[u] + graph[u][w]
                    path[w] = u
                    
    return path, dist


def getMinVertex(dist, selected):
    minv = -1
    mindist = INF
    
    for i in range(len(dist)):
        if not selected[i] and mindist > dist[i]:
            mindist = dist[i]
            minv = i
            
    return minv


start = 0
path, dist = dijkstra(g, vertex, start)

print(path)

for end in range(len(vertex)):
    if end != start:
        print("[최단경로 : %s -> %s, %d] %s" % (vertex[start], vertex[end], dist[end], vertex[end]), end = '')
        while path[end] != start:
            print(" <- %s" % vertex[path[end]], end = '')
            end = path[end]
        print(" <- %s" % vertex[path[end]])
        
```
